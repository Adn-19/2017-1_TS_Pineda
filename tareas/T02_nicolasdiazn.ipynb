# Tarea

Logistic Map : $$ x_{n+1} = rx_n(1-x_n)$$

```julia
function log_map(x,r,N,M)
    #N es el número de iteraciones
    #M será el número de iteraciones que deseamos ignorar.
    v = Float64[]
    for i in 1:N
        x=x*r*(1-x)
        push!(v,x)
    end
    v
    deleteat!(v,1:M)
end
```

```julia
function bifurcation(N,M) #Esta función nos hará el gráfico de bifurcación tomando distintos valores de r. 
    ff = linspace(1,4,N-M)
    oo = [log_map(0.2,j,N,M) for j in ff]
    plot(ff,oo)

xlabel("r")
ylabel("x")
title("Bifurcación")
grid("on")    
    
    
end

#Tomo un ejemplo
bifurcation(1000,100)

#Ésta ya está lista.
```

```julia
function logplot(r,N) #De forma complementaria, se incluye una función que graficará, para r fija, los distintos valores iniciales.
    M = 0
    ii = linspace(0,1,N)
    YY = [log_map(i,r,N,M) for i in ii]
    plot(ii,YY)
    
xlabel("x_0")
ylabel("x_n")
#title("Mapeo Logístico para r=3 fija")
grid("on")
end

#Ejemplo

logplot(3,1000)
title("Mapeo Logístico para r=3 fija")
#logplot(2,1000)
```
