```julia
using PyPlot
```

```julia
function xnuevo(r,xanterior)
    return r*xanterior*(1-xanterior)
end
```

```julia
grafica=Float64[];
ejexgrafica=Float64[];
a=0.5;
for r in 1:0.02:2.9
    datos=Float64[];
    for i in 1:1:100
        a=xnuevo(r,a);
        b=Float64[a];
        append!(datos,b);
    end
    #plot(1:100,datos)
    append!(grafica,Float64[datos[end]])
    #println(grafica)
    append!(ejexgrafica,Float64[r])
    #println(ejexgrafica)
end

for r in 2.9:0.02:3.4
    datos=Float64[];
    for i in 1:1:100
        a=xnuevo(r,a);
        b=Float64[a];
        append!(datos,b);
    end
    #plot(1:100,datos)
    append!(grafica,Float64[datos[end]])
    append!(grafica,Float64[datos[end-1]])
    #println(grafica)
    append!(ejexgrafica,Float64[r])
    append!(ejexgrafica,Float64[r])
    #println(ejexgrafica)
end

for r in 3.4:0.01:3.6
    datos=Float64[];
    for i in 1:1:100
        a=xnuevo(r,a);
        b=Float64[a];
        append!(datos,b);
    end
    #plot(1:100,datos)
    append!(grafica,Float64[datos[end]])
    append!(grafica,Float64[datos[end-1]])
    append!(grafica,Float64[datos[end-2]])
    append!(grafica,Float64[datos[end-3]])
    #println(grafica)
    append!(ejexgrafica,Float64[r])
    append!(ejexgrafica,Float64[r])
    append!(ejexgrafica,Float64[r])
    append!(ejexgrafica,Float64[r])
    #println(ejexgrafica)
end
```
